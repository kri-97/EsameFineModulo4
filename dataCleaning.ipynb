```python
# funzione che mi dice se l'autore è anche autore (1 o 0)
def cantante_autore(row):
    autori = row["autori"].split(",")
    interprete = row["cantante"]
    
    # normalizzo i nomi degli autori e del cantante
    autori = [autore.strip().lower() for autore in autori]
    interprete_clean = interprete.strip().lower()

    # questi gruppi sono anche gli autori
    gruppi_speciali = ["måneskin", "il volo", "stadio", "homo sapiens"]
    if interprete_clean in gruppi_speciali:
        return 1
    
    # caso di cantante con solo un nome (es. Olly)
    if len(interprete.split()) == 1:
        if interprete_clean in autori:
            return 1
    
    # caso di cantante con nome e cognome (es. Angelina Mango)
    # osservazione: nella colonna autori, i cantanti sono scritti inizialeNome.Cognome (es. A.Mango)
    elif len(interprete.split()) > 1:
        interprete_split = interprete.split()
        interprete_abbrev = f"{interprete_split[0][0].lower()}. {interprete_split[-1].lower()}"
        if interprete_abbrev in autori:
            return 1
    
    return 0
```

### Costruzione Dataset Finale

```python
# inizializzazione dizionario per assegnare id univoco ad ogni cantante e ad ogni canzone
cantanti_dict = {}
canzoni_dict = {}
cantante_id = 1
canzone_id = 1

# costruzione dei dizionari
for j, rank_row in rank.iterrows():

    # è stata fatta una normalizzazione del file excel del rank:
    # perchè gli interpreti venivano divisi a volte con "e" altre con "-" o ","
    # ho deciso di separare tutto con " - "

    interpreti = rank_row["Interprete"].split(" - ")
    for interprete in interpreti:
        # Puliamo il nome del cantante
        interprete_clean = interprete.strip()
        if interprete_clean not in cantanti_dict:
            cantanti_dict[interprete_clean] = cantante_id
            cantante_id += 1
    
    if rank_row["Canzone"] not in canzoni_dict:
        canzoni_dict[rank_row["Canzone"]] = canzone_id
        canzone_id += 1

# inizializzo il contenitore finale
# voglio avere per ogni interprete/artista tutte le canzoni cantate (se da solista o in gruppo)
finale = []

# uso un doppio for per unire direttamente tutte le colonne che mi servivano
for i, row in festival.iterrows():
    for j, rank_row in rank.iterrows():

        # simula un join
        if rank_row["anno"] == row["Anno"]:
            
            # cerco la canzone corrispondente nell"ultimo dataframe per prendere durata, popolarità e url
            # grazie al filtro precedente dell"anno, anche se il titolo di una canzone si presenta più volte, sono sicuro di prendere la "canzone giusta"
            canzone_info = spotify_songs[spotify_songs["Canzone"] == rank_row["Canzone"]]
            if not canzone_info.empty:
                durata = canzone_info["Durata (min)"].values[0]
                popolarita = canzone_info["Popolarità"].values[0]
                url_album = canzone_info["URL immagine album"].values[0]
            else:
                # Impostiamo valori di default se non trovata la canzone
                durata = None
                popolarita = None
                url_album = None

            # verifico se l"interprete è un gruppo (mi salvo un booleano) splittando il valore
            interpreti = rank_row["Interprete"].split(" - ")
            gruppo = len(interpreti) > 1
            
            # costruzione della lista finale
            # per ogni inteprete mi costruisco la riga
            for interprete in interpreti:
                interprete_clean = interprete.strip()
                if interprete_clean in cantanti_dict:
                    finale.append({
                        "idSanremo": i+1,
                        "annoSanremo": row["Anno"],
                        "periodoSanremo": row["Periodo"],
                        "sede": row["Sede"],
                        "presentatore": row["Presentatore"],
                        "partecipanti": row["Partecipanti"],
                        "vincitore": row["Vincitore"],
                        "cantanteID": cantanti_dict[interprete_clean],
                        "cantante": interprete_clean,
                        "posizione": rank_row["Posizione"],
                        "canzone": rank_row["Canzone"],
                        "canzoneID": canzoni_dict[rank_row["Canzone"]],
                        "tipo": "Non Solista" if gruppo else "Solista",
                        "durata": durata,
                        "popolarità": popolarita,
                        "autori": rank_row["Autori"],
                        "urlAlbum": url_album
                    })
    
finale_df = pd.DataFrame(finale)

#aggiungo colonna per il numero di partecipazioni e il numero di vittorie
finale_df["partecipazioni"] = finale_df.groupby("cantante")["idSanremo"].transform("nunique")
finale_df["vittorie"] = finale_df[finale_df["posizione"] == "1º"].groupby("cantante")["cantante"].transform("count")

#aggiungo colonna che mi dice quanti autori hanno scritto la canzone
finale_df["numero_autori"] = finale_df["autori"].apply(lambda x: len(x.split(",")))

# richiamo la funzione creata
finale_df["cantante_autore"] = finale_df.apply(cantante_autore, axis=1)
```

```python
# aggiungo l"ID del cantante nel dataframe genere
genere["cantanteID"] = genere["Interprete"].map(cantanti_dict)

# Ora uniamo finale_df con genere utilizzando la colonna "cantanteID"
finale_df = finale_df.merge(genere[["cantanteID", "Sesso"]], on="cantanteID", how="left")

finale_df = finale_df.drop_duplicates()
finale_df
```
